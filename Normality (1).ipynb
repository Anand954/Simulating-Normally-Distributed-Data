## **Simulating Normal Data :-**

```python
from numpy.random import seed
from numpy.random import normal
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro
seed(1)

#generating sample of 400 values that follow a normal distribution mean 0 and standard deviation 1
data = normal(loc=0, scale=1, size=400)

#observing first twenty values
data[0:20]
```

```python
#Finding the mean of above sample
np.mean(data)
```

```python
#Finding the standard deviation of above sample
np.std(data, ddof=1)
```

```python
#Plotting histogram
count, bins, ignored = plt.hist(data, 30)
plt.show()
```

```python
#Shapiro-Wilk test
shapiro(data)
```

```python
#create Q-Q plot with 45-degree line added to plot
import statsmodels.api as sm
fig = sm.qqplot(data, line='45')
plt.show()
```
