```python
from numpy.random import seed
from numpy.random import normal
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro
seed(1)

#generating sample of 400 values that follow a normal distribution 
#with mean 0 and standard deviation 1
data = normal(loc=0, scale=1, size=400)

#observing first ten values
data[0:10]
```

```python
#Finding the mean of above sample
np.mean(data)
```

```python
#Finding the standard deviation of above sample
np.std(data, ddof=1)
```

```python
#Plotting histogram
count, bins, ignored = plt.hist(data, 40)
plt.show()
```

```python
#Shapiro-Wilk test
shapiro(data)
```
